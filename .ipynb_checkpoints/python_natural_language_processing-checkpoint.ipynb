```python
import time
import nltk

# Run timer
start = time.time()

# Import the required resources (nltk package)
nltk.download('punkt')

# Read in the book text
filename = "/Users/dariamartinovskaya/Downloads/PLN/Chapter01/sherlock_holmes_1.txt"
with open(filename, "r", encoding="utf-8") as file:
    text = file.read()

# Replace newlines with spaces
text = text.replace("\n", " ")

# Initialize an nltk tikenizer
tokenizer = nltk.data.load("tokenizers/punkt/english.pickle")

# Divide the text into sentences
sentences = tokenizer.tokenize(text)

# The resulting list
for sentence in sentences:
    print(sentence)

# Count time
end = time.time()

print("")
print(f"Time spent: {end - start:.4f} sec")
```

```python
import time
import spacy

# Run timer
start = time.time()

# Read in the book text
filename = "/Users/dariamartinovskaya/Downloads/PLN/Chapter01/sherlock_holmes_1.txt"
file = open(filename, "r", encoding="utf-8")
text = file.read()

# Replace newlines with spaces
text = text.replace("\n", " ")

# Initialize the spaCy engine
nlp = spacy.load("en_core_web_sm")

# Divide the text into sentences
doc = nlp(text)
sentences = [sentence.text for sentence in doc.sents]

# Print all sentences
for sentence in sentences:
    print(sentence)

# Count time
end = time.time()

print("")
print(f"Time spent: {end - start:.4f} sec")
```

```python
# DIVIDING SENTENCES INTO WORDS - TOKENIZATION

import nltk

# Read in the book text
filename = "/Users/dariamartinovskaya/Downloads/PLN/Chapter01/sherlock_holmes_1.txt"
with open(filename, "r", encoding="utf-8") as file:
    text = file.read()

# Replace newlines with spaces
text = text.replace("\n", " ")

# Divide the text into words
words = nltk.tokenize.word_tokenize(text)

# Print output
for word in words:
    print(word)
```

```python
import spacy

# Read in the book text
filename = "/Users/dariamartinovskaya/Downloads/PLN/Chapter01/sherlock_holmes_1.txt"
with open(filename, "r", encoding="utf-8") as file:
    text = file.read()

# Replace newlines with spaces
text = text.replace("\n", " ")

# Initialize the spaCy engine using the English model
nlp = spacy.load("en_core_web_sm")

# Divide the text into sentences
doc = nlp(text)
words = [token.text for token in doc]

# Print output
for word in words:
    print(word)
```
